```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate universe variables
x = np.arange(0, 11, 1)

# Generate fuzzy membership functions
low = fuzz.trimf(x, [0, 0, 434])
medium = fuzz.trimf(x, [424-(424*0.25), 424,424+(424*0.25)])
high = fuzz.trimf(x, [434, 690, 690])

# Plot the membership functions
plt.figure()
plt.plot(x, low, 'b', linewidth=1.5, label='Low')
plt.plot(x, medium, 'g', linewidth=1.5, label='Medium')
plt.plot(x, high, 'r', linewidth=1.5, label='High')
plt.title('Fuzzy membership functions')
plt.legend()
plt.show()
```
